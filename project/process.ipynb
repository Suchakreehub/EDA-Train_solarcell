```python
import joblib
import pandas as pd

model_path = "randomfor.pkl"  
rf_model = joblib.load(model_path)

# โหลดข้อมูลใหม่ (เปลี่ยนเป็นไฟล์จริงที่ต้องการใช้พยากรณ์)
data_path = "newest _data.csv"  # หรือ "new_data.csv"
df = pd.read_csv(data_path, index_col=0) # ใช้ pd.read_csv(data_path) ถ้าเป็นไฟล์ CSV
df = df[df['Irradiance'] > 0].reset_index(drop=True)

# ตรวจสอบคอลัมน์ที่ต้องใช้ในการพยากรณ์
features = ['PV_input_current(A)', 'PV_input_voltage(V)', 'Ambient_temp', 
            'Irradiance', 'PV_temp']

df_test = df[features]  # เลือกเฉพาะคอลัมน์ที่ใช้ในโมเดล

# ใช้โมเดลที่โหลดมาทำนาย
df["Anomaly_Prediction"] = rf_model.predict(df_test)

# แสดงผลลัพธ์
print(df.head()) 
```

```python
anomaly_counts = df.groupby("PV_no.")["Anomaly_Prediction"].value_counts()  # นับ anomaly แต่ละ PV_no
total_counts = df["PV_no."].value_counts().sort_index()
anomaly_total = df[df["Anomaly_Prediction"].isin([1, 2, 3])].groupby("PV_no.")["Anomaly_Prediction"].count()
anomaly_percentage = (anomaly_counts.div(total_counts, axis=0)) * 100
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# สร้าง Mapping ของ Label ใหม่
label_mapping = {
    0: "Normal",
    1: "Volatage  Anomaly",
    2: "Ampare Anomaly",
    3: "Volt&Amp Anomaly"
}

# ตรวจสอบว่ามีคอลัมน์ที่ต้องใช้หรือไม่
if "PV_no." not in df.columns or "Anomaly_Prediction" not in df.columns:
    raise ValueError("Dataframe must contain 'PV_no.' and 'Anomaly_Prediction' columns.")

# แปลงค่า Anomaly_Prediction เป็น int (เผื่อเป็น float)
df["Anomaly_Prediction"] = df["Anomaly_Prediction"].astype(int)

# เปลี่ยนค่าตัวเลขเป็น Label ที่อ่านง่าย
df["Anomaly_Label"] = df["Anomaly_Prediction"].map(label_mapping)

# นับจำนวนค่าของ Anomaly_Label ในแต่ละ PV_no.
df_pie = df.groupby("PV_no.")["Anomaly_Label"].value_counts().unstack()

# พล็อต Pie Chart พร้อม Label ใหม่
for pv_no, data in df_pie.iterrows():
    plt.figure(figsize=(6, 6))
    data.dropna(inplace=True)  # ลบค่าที่เป็น NaN
    plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Anomaly Distribution for PV_no. {pv_no}')
    plt.show()
```

```python
# โหลดโมเดลที่บันทึกไว้
from tensorflow.keras.models import load_model
import joblib

# โหลดโมเดล LSTM
model_lstm = load_model("lstm_anomaly_model.h5")

# โหลด Scaler ที่ใช้กับข้อมูลเก่า
scaler = joblib.load("scaler.pkl")

print("โหลดโมเดลและ Scaler สำเร็จ!")
```

```python
import numpy as np
import pandas as pd



# แปลงคอลัมน์ Date เป็น datetime
df_predic['Date'] = pd.to_datetime(df_predic['Date'], errors='coerce')

# เลือกเฉพาะ PV_no. 4 (ถ้าต้องการข้อมูลของ PV No. 4 เท่านั้น)
new_data_pv4 = df_predic[df_predic['PV_no.'] == 4].copy()

# สรุปจำนวน anomaly ต่อวัน
new_anomaly_pv4 = new_data_pv4.groupby('Date')['Anomaly_Label'].sum().reset_index()

# ใช้ Scaler ที่เคยบันทึกไว้ในการแปลงข้อมูลใหม่
new_anomaly_pv4_scaled = scaler.transform(new_anomaly_pv4[['Anomaly_Label']])

# ใช้ข้อมูล 10 วันสุดท้ายของชุดข้อมูลใหม่เพื่อทำการพยากรณ์
seq_length = 10
if len(new_anomaly_pv4_scaled) < seq_length:
    raise ValueError("ข้อมูลใหม่มีไม่เพียงพอ ต้องมีอย่างน้อย 10 วันสำหรับการพยากรณ์")

# ดึง 10 วันสุดท้ายเพื่อใช้เป็นอินพุตให้โมเดล
input_sequence = new_anomaly_pv4_scaled[-seq_length:]
input_sequence = np.expand_dims(input_sequence, axis=0)  # ปรับขนาดให้เหมาะสมกับโมเดล LSTM

# ใช้โมเดลทำการพยากรณ์ anomaly ในวันถัดไป
predicted_anomaly_scaled = model_lstm.predict(input_sequence)
predicted_anomaly = scaler.inverse_transform(predicted_anomaly_scaled)

print(f"จำนวน anomaly ที่คาดการณ์สำหรับวันถัดไป: {predicted_anomaly[0][0]:.2f}")
```

```python
future_steps = 7  # จำนวนวันที่ต้องการพยากรณ์
forecast_results = []

current_seq = input_sequence.copy()

for _ in range(future_steps):
    predicted = model_lstm.predict(current_seq)
    forecast_results.append(predicted[0, 0])

    # อัปเดต sequence โดยเลื่อนค่ามา 1 วัน และเพิ่มค่าที่พยากรณ์ได้เข้าไป
    current_seq = np.roll(current_seq, -1, axis=1)
    current_seq[0, -1, 0] = predicted[0, 0]

# แปลงค่ากลับไปยังสเกลเดิม
forecast_results = scaler.inverse_transform(np.array(forecast_results).reshape(-1, 1)).flatten()

# สร้างวันที่สำหรับพยากรณ์
future_dates = pd.date_range(start=new_anomaly_pv4['Date'].iloc[-1] + pd.Timedelta(days=1), periods=future_steps)

# แสดงผลลัพธ์เป็น DataFrame
forecast_df = pd.DataFrame({'Date': future_dates, 'Predicted_Anomaly': forecast_results})
print(forecast_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(new_anomaly_pv4['Date'], new_anomaly_pv4['Anomaly_Label'], label="Actual Anomalies (PV No. 4)", marker='o')
plt.plot(future_dates, forecast_results, label="Forecasted Anomalies (LSTM, PV No. 4)", marker='o', linestyle='dashed', color='red')
plt.xlabel("Date")
plt.ylabel("Anomaly Count")
plt.title("Anomaly Count Forecast using LSTM")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
